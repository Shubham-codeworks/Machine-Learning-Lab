# Discovering Kepler’s third law from planetary data

## Goal
In this data-inspired project, we illustrate how Kepler’s Third Law of Planetary Motion can be discovered using regression modeling to fit a power model to real planetary data obtained from NASA. The power model can be linearized, thus we can use linear regression to fit the model parameters to the data, but we also show how a non-linear regression can be implemented using the Python programming language.

## Background
### Kepler's law of harmomnies
Kepler’s third law is an elegant example of a power law between the period of an orbiting body and its average distance to the object it orbits.
*  Kepler’s laws can be derived from Newton’s laws of motion and his law of gravity.
*  Kepler’s laws allow us to use them as tools to do space explorations and make predictions.
*  Kepler’s third law allows us to find the most efficient trajectory from Earth to Mars, which is called the Hohmann transfer orbit. Kepler’s third law implies that this orbit has to be an ellipse that touches both orbits and has the shortest possible semimajor axis.


## Planetary data 
#### Source: NASA’s Lunar and Planetary Science Division

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import statsmodels.formula.api as smf
```

```python
df = pd.DataFrame({"Planet" : ["Mercury","Venus","Earth","Mars","Jupiter","Saturn","Uranus","Neptune","Pluto"], 
                   "Distance (in 10^6 km)" : [57.9,108.2,149.6,227.9,778.6,1433.5,2872.5,4495.1,5906.4],
                   "Period (in days)" : [88,224.7,365.2,687,4331,10747,30589,59800,90560]})
df
```

## Normalized planetary data 
* The planetary data were derived from the original NASA data by doing data transformations that consists of normalizing the distances and the orbital periods in Earth’s units (1AU = 1.496 × 108 km).
* We use the normalized planetary data to find a relationship between the average distance from the Sun, given
in AU, and the orbital period, given in Earth’s years.
* For the model function, we use a power model with two parameters $α$ and $β$:                                                                                                                     $T = αD^{β}$
* Here T is the orbital period in Earth’s years and D is the average distance from the Sun in AU. 

```python
def f(x):
    return x / df["Distance (in 10^6 km)"][2]

def g(y):
    return y / df["Period (in days)"][2]

def f1(x):
    return np.log(x)

def g1(y):
    return np.log(y)

df["Distance D [AU]"] = df["Distance (in 10^6 km)"].apply(f)
df["Period T [Earth’s Years]"] = df["Period (in days)"].apply(g)
df["ln(D)"] = df["Distance D [AU]"].apply(f1)
df["ln(T)"] = df["Period T [Earth’s Years]"].apply(g1)

df[["Planet", "Distance D [AU]", "Period T [Earth’s Years]", "ln(D)", "ln(T)"]]
```

## Visualizing the normalized planetary data

```python
x, y = df["Distance D [AU]"], df["Period T [Earth’s Years]"]
plt.figure(figsize=(10,5))
plt.scatter(x, y, color='r')
plt.title("The Normalised Planetary Data")
plt.xlabel("Distance D [AU]")
plt.ylabel("Period T [Earth’s Years]")
plt.grid()
plt.show()
```

## Fitting the model with linear least squares
* We can linearize the power model  $T = αD^{β}$ and then apply linear least squares to fit the model parameters.
* The linearization consists of taking the logarithm on both sides: $ln(T) = ln(α) + β ln(D)$ 
* Now, we have a linear model with respect to the unknown parameters $γ = ln(α)$ and $β$. The response variable
is now $y = ln(T)$, and the predictor is $x = ln(D).$
* Thus, the linearized model can be written as:
$y = γ + βx $

```python
dataDT = pd.DataFrame({"D" : df["Distance D [AU]"], "T" : df["Period T [Earth’s Years]"]})
dataDT
```

```python
model = smf.ols('np.log(T) ~ np.log(D)', data=dataDT).fit()
print(model.summary())
```

## Fitting a linear model without an intercept
* The intercept (the constant term γ) is not statistically significant since its p-value is about 81%. Thus, we
can update our linear model to one with $γ = 0 $ $(α = 1)$:
* $y = βx ⇐⇒ T = D^β$


```python
updated_model = smf.ols('np.log(T) ~ 0 + np.log(D)', data=dataDT).fit()
print(updated_model.summary())
```

## Kepler’s third law rediscovered
* Thus, we can conclude that β is very likely to be 1.5 in a physical model of our planetary system.
* Dimensional analysis would also confirm the value of 1.5.
* The fitted model then becomes: $T = D^{3/2} ⇐⇒ T^2 = D^3$
* This is what Kepler discovered in 1619 with his Third Law, about 400 years ago.


```python
x, y = df["Distance D [AU]"], df["Period T [Earth’s Years]"]
plt.figure(figsize = (10, 5))
plt.scatter(x, y, color='r')
l, m = dataDT["D"], dataDT["T"]
plt.plot(l, m)
plt.title("The fitted power model $T = D^{3/2}$ (In Blue) and planetary data (In Red)")
plt.xlabel("Distance D [AU]")
plt.ylabel("Period T [Earth’s Years]")
plt.grid()
plt.show()
```
