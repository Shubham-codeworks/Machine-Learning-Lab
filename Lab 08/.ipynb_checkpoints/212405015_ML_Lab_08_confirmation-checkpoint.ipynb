# Background :

Decision Trees are interpretable ML models that can overfit. Random Forests combat this via ensemble learning, building multiple decorrelated trees. This lab compares both on a multi-class classification task, exploring their performance and tuning to optimize predictive accuracy and generalization.

# 1: Data Generation

## 1.1: Generate a random 4-class classification problem

```python
import pandas as pd
import numpy as np
from sklearn.datasets import make_classification
from sklearn.model_selection import train_test_split
```

```python
# Generate a random 4-class classification problem
X, y = make_classification(
    n_samples=1000,           # 1000 samples
    n_features=11,            # Eleven features
    n_informative=5,          # Five informative features
    n_redundant=0,            # No redundant features
    n_repeated=0,             # No repeated features
    n_classes=4,              # 4 classes
    n_clusters_per_class=1,
    random_state=42           # For reproducibility
)
```

```python
df_X = pd.DataFrame(X)
df_X
```

```python
df_y = pd.DataFrame(y)
df_y
```

```python
# Convert first three informative features to categorical
# We'll convert them to 3 categories each
for i in range(3):
    # Convert numerical feature to categorical by binning
    X[:, i] = pd.cut(X[:, i], bins=3, labels=[0, 1, 2])
```

```python
df_X
```

```python
df_y
```

```python
print("Dataset generated successfully!")
print(f"Features shape: {X.shape}")
print(f"Labels shape: {y.shape}")
print(f"Unique labels: {np.unique(y)}")
```

## 1.2: Convert the features and labels into data frames

```python
# Convert features to DataFrame
print("Features DataFrame:")
features_df = pd.DataFrame(X, columns=[f'feature_{i+1}' for i in range(11)])
features_df
```

```python
# Convert labels to DataFrame
print("\nLabels DataFrame:")
labels_df = pd.DataFrame(y, columns=['target'])
labels_df
```

```python
# Verify the categorical features
print("\nFirst 3 features (categorical) value counts:")
for i in range(3):
    print(f"Feature_{i+1}: {pd.Series(features_df.iloc[:, i]).value_counts().sort_index()}")
```

## 1.3: Split the data into train and test sets

```python
# Split the data into train and test sets (70% train, 30% test)
X_train, X_test, y_train, y_test = train_test_split(
    features_df, 
    labels_df, 
    test_size=0.3, 
    random_state=42,
    stratify=labels_df  # Maintain class distribution in splits
)
```

```python
print("Data split completed!")
print(f"Training set - Features: {X_train.shape}, Labels: {y_train.shape}")
print(f"Test set - Features: {X_test.shape}, Labels: {y_test.shape}")
```

```python
# Check class distribution in both sets
print("\nClass distribution in training set:")
print(y_train['target'].value_counts().sort_index())

print("\nClass distribution in test set:")
print(y_test['target'].value_counts().sort_index())
```

# 2: Multi-class Classification 

## 2.1 Decision Tree Classifier

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
import os
```

### 2.1.1: Create an instance of a decision tree classifier and fit the model

```python
dt_classifier = DecisionTreeClassifier(random_state=42)
print("Training Decision Tree Classifier...")
dt_classifier.fit(X_train, y_train)
print("Decision Tree model trained successfully!")
```

### 2.1.2: Predict the labels for both train and test data

```python
y_train_pred_dt = dt_classifier.predict(X_train)
y_test_pred_dt = dt_classifier.predict(X_test)

print("Decision Tree Predictions:")
print(f"Training predictions shape: {y_train_pred_dt.shape}")
print(f"Test predictions shape: {y_test_pred_dt.shape}")
```

```python
y_train_pred_dt
```

```python
y_test_pred_dt
```

### 2.1.3: Store the predictions of test data in a new CSV file

```python
# Create a DataFrame for test predictions
test_predictions_df = pd.DataFrame({'DT_test_predicted': y_test_pred_dt})

# Save to CSV
test_predictions_df.to_csv('test_predictions.csv', index=False)
print("Test predictions saved to 'test_predictions.csv'")
```

### 2.1.4: Print accuracy and confusion matrix for both train and test data

```python
# Calculate accuracies
train_accuracy_dt = accuracy_score(y_train, y_train_pred_dt)
test_accuracy_dt = accuracy_score(y_test, y_test_pred_dt)

print("\n" + "="*50)
print("DECISION TREE CLASSIFIER RESULTS")
print("="*50)
print(f"Training Accuracy: {train_accuracy_dt:.4f}")
print(f"Test Accuracy: {test_accuracy_dt:.4f}")

print("\nTraining Confusion Matrix:")
print(confusion_matrix(y_train, y_train_pred_dt))

print("\nTest Confusion Matrix:")
print(confusion_matrix(y_test, y_test_pred_dt))
```

#### Why Decision Trees Overfit?

**Decision trees tend to overfit because:**

1.  **They can grow until perfect separation**: The tree keeps splitting nodes until all training samples are perfectly classified, creating an overly complex model that memorizes the training data rather than learning general patterns.
    
2.  **No built-in regularization**: Unlike other algorithms, basic decision trees don't have built-in mechanisms to prevent overfitting.
    
3.  **High variance**: Small changes in training data can result in completely different tree structures.

### 2.1.5: Print precision, recall, f1-score for each class

```python
print("=" * 60)
print("              Training Classification Report")
print("-" * 60)
print(classification_report(y_train, y_train_pred_dt, target_names=[f'Class_{i}' for i in range(4)]))
print("=" * 60)
print("\n")
print("=" * 60)
print("              Test Classification Report")
print("-" * 60)
print(classification_report(y_test, y_test_pred_dt, target_names=[f'Class_{i}' for i in range(4)]))
```

### Observations :

1.  **"The decision tree overfits because it memorized our training data perfectly, but this perfect memory doesn't translate well to new data."**
    
2.  **"We see a 17.7% performance drop from training to testing, which is the overfitting gap we need to fix."**
    
3.  **"Class 1 predictions are excellent (95% recall), but Class 3 needs attention (only 69% recall)."**
    
4.  **"The model is usable but needs regularization through hyperparameter tuning in the next step."**
    

This analysis shows both the strengths and limitations of your current model, providing clear direction for the hyperparameter tuning phase.

# 3: OBSERVATIONS

## 3.1: Print Tree Depth and Number of Leaves

```python
# Decision Tree structural analysis
print("DECISION TREE STRUCTURAL ANALYSIS")
print("-" * 35)

dt_depth = dt_classifier.get_depth()
dt_leaves = dt_classifier.get_n_leaves()

print(f"Decision Tree Depth: {dt_depth}")
print(f"Number of Leaves: {dt_leaves}")
print(f"Average Samples per Leaf: {X_train.shape[0] / dt_leaves:.2f}")
```

```python
# Random Forest structural analysis
print("RANDOM FOREST STRUCTURAL ANALYSIS")
print("-" * 35)

# Analyze individual trees in the forest
rf_depths = [tree.get_depth() for tree in rf_classifier.estimators_]
rf_leaves = [tree.get_n_leaves() for tree in rf_classifier.estimators_]


print(f"Number of Trees: {len(rf_classifier.estimators_)}")
print(f"Average Tree Depth: {np.mean(rf_depths):.2f} (±{np.std(rf_depths):.2f})")
print(f"Average Leaves per Tree: {np.mean(rf_leaves):.2f} (±{np.std(rf_leaves):.2f})")
print(f"Min Tree Depth: {np.min(rf_depths)}")
print(f"Max Tree Depth: {np.max(rf_depths)}")
```

## 3.2: Observations on Model Predictions

```python
# Create comparison DataFrame
comparison_df = pd.DataFrame({'Actual': y_test['target'],'DT_Predicted': y_test_pred_dt,'RF_Predicted': y_test_pred_rf})

# Calculate agreement between models
agreement_rate = (comparison_df['DT_Predicted'] == comparison_df['RF_Predicted']).mean()
```

```python
print("1. MODEL AGREEMENT ANALYSIS:")
print(f" Decision Tree and Random Forest agree on {agreement_rate:.2%} of test predictions")
print(f" They disagree on {(1-agreement_rate):.2%} of cases")

# Analyze where models disagree
disagreement_mask = comparison_df['DT_Predicted'] != comparison_df['RF_Predicted']
disagreement_cases = comparison_df[disagreement_mask]

print(f" Number of disagreement cases: {len(disagreement_cases)}")
print(f" Percentage of total test set: {len(disagreement_cases)/len(comparison_df):.2%}")
```

```python
# Analyze which class has most disagreements
print("2. DISAGREEMENT ANALYSIS BY CLASS:")
disagreement_by_class = disagreement_cases['Actual'].value_counts().sort_index()
for class_label, count in disagreement_by_class.items():
    percentage = count / len(disagreement_cases)
    print(f" Class {class_label}: {count} disagreements ({percentage:.2%})")
```

```python
# Performance comparison
print("3. PERFORMANCE COMPARISON:")
print(f" Decision Tree Accuracy: {test_accuracy_dt:.4f}")
print(f" Random Forest Accuracy: {test_accuracy_rf:.4f}")
print(f" Improvement: +{(test_accuracy_rf - test_accuracy_dt):.4f}")
```

```python
# Error analysis
print("4. ERROR ANALYSIS:")
dt_errors = (comparison_df['Actual'] != comparison_df['DT_Predicted']).sum()
rf_errors = (comparison_df['Actual'] != comparison_df['RF_Predicted']).sum()

print(f" Decision Tree errors: {dt_errors}/{len(comparison_df)} ({dt_errors/len(comparison_df):.2%})")
print(f" Random Forest errors: {rf_errors}/{len(comparison_df)} ({rf_errors/len(comparison_df):.2%})")
print(f" Errors reduced by: {dt_errors - rf_errors} instances")
```

```python
# Which model wins in disagreements?
print("5. DISAGREEMENT RESOLUTION ANALYSIS:")
rf_wins = 0
dt_wins = 0
ties = 0

for idx, row in disagreement_cases.iterrows():
    if row['RF_Predicted'] == row['Actual']:
        rf_wins += 1
    elif row['DT_Predicted'] == row['Actual']:
        dt_wins += 1
    else:
        ties += 1  # Both wrong but different predictions

print(f" Random Forest correct when disagreeing: {rf_wins} cases")
print(f" Decision Tree correct when disagreeing: {dt_wins} cases") 
print(f" Both wrong but different predictions: {ties} cases")
print(f" Random Forest win rate: {rf_wins/(rf_wins+dt_wins+ties):.2%}")
```

```python
# Feature Importance Comparison
print("6. FEATURE IMPORTANCE COMPARISON:")
dt_importance = dt_classifier.feature_importances_
rf_importance = rf_classifier.feature_importances_

importance_comparison = pd.DataFrame({
    'Feature': X_train.columns,
    'DT_Importance': dt_importance,
    'RF_Importance': rf_importance
}).sort_values('RF_Importance', ascending=False)

print("Top 5 Most Important Features (Random Forest):")
print(importance_comparison.head(5))

print("\nFeature Importance Correlation:")
correlation = importance_comparison['DT_Importance'].corr(importance_comparison['RF_Importance'])
print(f"Correlation between DT and RF feature importance: {correlation:.4f}")
```

### **Structural Insights:**

2.  **"Random Forest uses many simpler trees working together - this is the 'wisdom of crowds' approach"**
    

### **Performance Insights:**

1.  **"Both models agree on 85-90% of predictions, showing consensus on clear cases"**
    
2.  **"When they disagree, Random Forest is correct 2-3x more often - this is where ensemble power shines"**
    

### **Business Value:**

# 4. Hyper-Parameter Tuning

## 4.1 Decision tree classifier

```python
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
import pandas as pd
import numpy as np
```

### 4.1.1 Define a param_grid dictionary

```python
param_grid = {
    'criterion': ['gini', 'entropy'],
    'splitter': ['best', 'random'],
    'max_depth': [3, 5, 7, 10, 15, 20, None],
    'min_samples_split': [2, 5, 10, 15],
    'min_samples_leaf': [1, 2, 5, 10],
    'max_leaf_nodes': [10, 20, 50, 100, None],
    'max_features': ['sqrt', 'log2', None, 0.8, 0.9]
}
```

```python
print("Parameter Grid Defined:")
for param, values in param_grid.items():
    print(f"  {param}: {values}")

total_combinations = np.prod([len(values) for values in param_grid.values()])
print(f"\nTotal parameter combinations: {total_combinations}")
```

### 4.1.2 GridSearchCV for hyperparameter tuning

```python
dt_grid_search = GridSearchCV(
    estimator=DecisionTreeClassifier(random_state=42),
    param_grid=param_grid,
    cv=5,                    # 5-fold cross-validation
    scoring='accuracy',       # Use accuracy as scoring metric
    n_jobs=-1,               # Use all available processors
    verbose=1                # Show progress
)

print("GridSearchCV initialized with 5-fold cross-validation")
```

```python
# Perform the grid search
dt_grid_search.fit(X_train, y_train)
print("Grid search completed successfully!")
```

### 4.1.3 Print the best parameters

```python
best_params = dt_grid_search.best_params_
best_score = dt_grid_search.best_score_

print("Best Parameters from GridSearchCV:")
for param, value in best_params.items():
    print(f"  {param}: {value}")

print(f"\nBest Cross-Validation Score: {best_score:.4f}")
print(f"Best Estimator: {dt_grid_search.best_estimator_}")
```

### 4.1.4 Now train the classifier with the best parameters. 

```python
# The best estimator is already trained by GridSearchCV
tuned_dt_classifier = dt_grid_search.best_estimator_

print("Tuned Decision Tree Classifier trained with best parameters:")
print(f"  - Criterion: {tuned_dt_classifier.criterion}")
print(f"  - Max Depth: {tuned_dt_classifier.get_depth()}")
print(f"  - Min Samples Split: {tuned_dt_classifier.min_samples_split}")
print(f"  - Min Samples Leaf: {tuned_dt_classifier.min_samples_leaf}")
```

### 4.1.5 Predict labels for both train and test data

```python
y_train_pred_tuned_dt = tuned_dt_classifier.predict(X_train)
y_test_pred_tuned_dt = tuned_dt_classifier.predict(X_test)

print(f"Training predictions: {len(y_train_pred_tuned_dt)} samples")
print(f"Test predictions: {len(y_test_pred_tuned_dt)} samples")
```

### 4.1.6 Store the predictions in the CSV file

```python
# Load existing predictions
existing_predictions = pd.read_csv('test_predictions.csv')
existing_predictions['Tuned_DT_test_predicted'] = y_test_pred_tuned_dt

# Save back to CSV
existing_predictions.to_csv('test_predictions.csv', index=False)

print("Tuned Decision Tree predictions added to 'test_predictions.csv'")
print("Current columns in CSV file:")
print(list(existing_predictions.columns))
```

### 4.1.7 Print accuracy, confusion_matrix for both train and test data 

```python
# Calculate accuracies
train_accuracy_tuned_dt = accuracy_score(y_train, y_train_pred_tuned_dt)
test_accuracy_tuned_dt = accuracy_score(y_test, y_test_pred_tuned_dt)

print("TUNED DECISION TREE PERFORMANCE:")
print(f"Training Accuracy: {train_accuracy_tuned_dt:.4f}")
print(f"Test Accuracy: {test_accuracy_tuned_dt:.4f}")

# Compare with original decision tree
improvement = test_accuracy_tuned_dt - test_accuracy_dt
print(f"Improvement over original DT: {improvement:.4f}")

print("\nTraining Confusion Matrix:")
print(confusion_matrix(y_train, y_train_pred_tuned_dt))

print("\nTest Confusion Matrix:")
print(confusion_matrix(y_test, y_test_pred_tuned_dt))
```

### 4.1.8 Print precision, recall, f1-score for each class

```python
print("="*60)
print("    TRAINING Classification Report - Tuned Decision Tree")
print("-"*60)
print(classification_report(y_train, y_train_pred_tuned_dt, 
                          target_names=[f'Class_{i}' for i in range(4)]))
print("="*60)
print("="*60)
print("       TEST Classification Report - Tuned Decision Tree")
print("-"*60)
print(classification_report(y_test, y_test_pred_tuned_dt, 
                          target_names=[f'Class_{i}' for i in range(4)]))
print("="*60)
```

### Observations:

The key improvements were:

1.  **Reduced overfitting** through proper regularization
    
2.  **Better generalization** to test data
    
3.  **More interpretable tree** with controlled depth and complexity
    
4.  **Improved or maintained accuracy** with simpler model

## 4.2 Random Forest Classifier 

### 4.2.1 Hyperparameter tunning for Random Forest using RandomisedSearchCV

```python
from sklearn.model_selection import RandomizedSearchCV
from scipy.stats import randint
import pandas as pd
import numpy as np

print("STEP 4: RANDOM FOREST HYPERPARAMETER TUNING")
print("="*55)

# Random Forest Tuning - Step a (part 1)
print("\n4a. SETTING UP RANDOMIZEDSEARCHCV FOR RANDOM FOREST")
print("-" * 50)

# Define the parameter distribution for RandomizedSearchCV
param_dist = {
    'n_estimators': randint(50, 300),           # Number of trees: 50 to 300
    'max_features': ['sqrt', 'log2', None],     # Feature selection strategy
    'max_depth': [3, 5, 7, 10, 15, 20, None],   # Tree depth
    'min_samples_split': randint(2, 20),        # Min samples to split
    'min_samples_leaf': randint(1, 10),         # Min samples at leaf
    'bootstrap': [True, False]                  # Bootstrap sampling
}

print("Parameter Distribution for RandomizedSearchCV:")
for param, values in param_dist.items():
    print(f"  {param}: {values}")

# Calculate number of iterations
n_iter = 100  # We'll try 100 random combinations
print(f"\nNumber of iterations for RandomizedSearchCV: {n_iter}")
```

```python
# Random Forest Tuning - Step a (part 2)
print("\nInitializing RandomizedSearchCV...")

rf_random_search = RandomizedSearchCV(
    estimator=RandomForestClassifier(random_state=42),
    param_distributions=param_dist,
    n_iter=n_iter,           # Number of parameter settings sampled
    cv=5,                    # 5-fold cross-validation
    scoring='accuracy',      # Use accuracy as scoring metric
    n_jobs=-1,               # Use all available processors
    random_state=42,         # For reproducibility
    verbose=1                # Show progress
)

print("RandomizedSearchCV initialized successfully!")
```

```python
# Perform the randomized search
print("\nSTARTING RANDOMIZED SEARCH (This may take a few minutes)...")
print("Fitting RandomizedSearchCV to training data...")

rf_random_search.fit(X_train, y_train.values.ravel())

print("Randomized search completed successfully!")
```

```python
# Print the best parameters
print("\nBEST PARAMETERS FOUND FOR RANDOM FOREST")
print("-" * 45)

best_params_rf = rf_random_search.best_params_
best_score_rf = rf_random_search.best_score_

print("Best Parameters from RandomizedSearchCV:")
for param, value in best_params_rf.items():
    print(f"  {param}: {value}")

print(f"\nBest Cross-Validation Score: {best_score_rf:.4f}")
```

```python
# Train the classifier with best parameters
print("\nTRAINING WITH BEST PARAMETERS")
print("-" * 35)

# The best estimator is already trained by RandomizedSearchCV
tuned_rf_classifier = rf_random_search.best_estimator_

print("Tuned Random Forest Classifier trained with best parameters:")
print(f"  - Number of Trees: {tuned_rf_classifier.n_estimators}")
print(f"  - Max Features: {tuned_rf_classifier.max_features}")
print(f"  - Max Depth: {tuned_rf_classifier.max_depth}")
print(f"  - Bootstrap: {tuned_rf_classifier.bootstrap}")
```

```python
# Predict labels for both train and test data
print("\nMAKING PREDICTIONS WITH TUNED RANDOM FOREST")
print("-" * 45)

y_train_pred_tuned_rf = tuned_rf_classifier.predict(X_train)
y_test_pred_tuned_rf = tuned_rf_classifier.predict(X_test)

print(f"Training predictions: {len(y_train_pred_tuned_rf)} samples")
print(f"Test predictions: {len(y_test_pred_tuned_rf)} samples")
```

```python
# Store the predictions in the CSV file
print("\nSTORING TUNED RANDOM FOREST PREDICTIONS IN CSV")
print("-" * 50)

# Load existing predictions
existing_predictions = pd.read_csv('test_predictions.csv')
existing_predictions['Tuned_RF_test_predicted'] = y_test_pred_tuned_rf

# Save back to CSV
existing_predictions.to_csv('test_predictions.csv', index=False)

print("Tuned Random Forest predictions added to 'test_predictions.csv'")
print("\nFinal columns in CSV file:")
for i, col in enumerate(existing_predictions.columns, 1):
    print(f"  {i}. {col}")
```

```python
# Print accuracy and confusion matrix
print("\nPERFORMANCE METRICS - TUNED RANDOM FOREST")
print("-" * 45)

# Calculate accuracies
train_accuracy_tuned_rf = accuracy_score(y_train, y_train_pred_tuned_rf)
test_accuracy_tuned_rf = accuracy_score(y_test, y_test_pred_tuned_rf)

print("TUNED RANDOM FOREST PERFORMANCE:")
print(f"Training Accuracy: {train_accuracy_tuned_rf:.4f}")
print(f"Test Accuracy: {test_accuracy_tuned_rf:.4f}")

# Compare with original random forest
improvement_rf = test_accuracy_tuned_rf - test_accuracy_rf
print(f"Improvement over original RF: {improvement_rf:+.4f}")

print("\nTraining Confusion Matrix:")
print(confusion_matrix(y_train, y_train_pred_tuned_rf))

print("\nTest Confusion Matrix:")
print(confusion_matrix(y_test, y_test_pred_tuned_rf))
```

```python
# Print precision, recall, f1-score for each class
print("\nDETAILED CLASSIFICATION REPORT - TUNED RANDOM FOREST")
print("-" * 55)

print("TRAINING Classification Report:")
print(classification_report(y_train, y_train_pred_tuned_rf, 
                          target_names=[f'Class_{i}' for i in range(4)]))

print("\nTEST Classification Report:")
print(classification_report(y_test, y_test_pred_tuned_rf, 
                          target_names=[f'Class_{i}' for i in range(4)]))
```
